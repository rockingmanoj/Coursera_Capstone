Q2. Given that this package can be very unreliable, in case you are not able to get the geographical coordinates of the neighborhoods using the Geocoder package, here is a link to a csv file that has the geographical coordinates of each postal code: http://cocl.us/Geospatial_data

Use the Geocoder package or the csv file to create the following dataframe

```
df_location=pd.read_csv('http://cocl.us/Geospatial_data')
```

```
df_location
```

```
print(df_location[df_location["Postal Code"]=='M1G'].iloc[0,1])
print(df_location[df_location['Postal Code']=="M1G"].iloc[0,2])
```

```
df_location[df_location["Postal Code"]=='M1G'].iloc[0,1].dtype
```

```
df_geo = pd.DataFrame(columns=['Latitude','Longitude'])
```

```
df_geo
```

```
for x in df_can_group.Postcode:
  Lat = df_location[df_location["Postal Code"]== x].iloc[0,1]
  Long = df_location[df_location["Postal Code"]== x].iloc[0,2]
  data = pd.DataFrame(data = [[Lat, Long]], columns=['Latitude','Longitude'])
  df_geo = pd.DataFrame(df_geo.append(data))
  df_geo = df_geo.reset_index(drop=True)
```

```
df_geo
```

```
df_merge = pd.concat([df_can_group, df_geo], axis=1) 
```

```
df_merge
```

Q3. Explore and cluster the neighborhoods in Toronto. You can decide to work with only boroughs that contain the word Toronto and then replicate the same analysis we did to the New York City data.

```
import folium
latitude = df_merge.Latitude.mean()
longitude = df_merge.Longitude.mean()
```

```
venues_map = folium.Map(location=[latitude, longitude], zoom_start=10) 

folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    popup = 'Ecco',
    fill = True,
    color = 'green',
    fill_color = 'yellow',
    fill_opacity=0.6
    ).add_to(venues_map)
```

```
df_merge.Latitude
df_merge.Longitude
df_merge.Neighbourhood
```

```
for lat, lng, label in zip(df_merge.Latitude, df_merge.Longitude, df_merge.Neighbourhood):
  folium.CircleMarker(
      [lat, lng],
      radius=5,
      popup = label,
      fill = True,
      color = 'blue',
      fill_color='blue',
      fill_opacity=0.6
  ).add_to(venues_map)
  venues_map
```
